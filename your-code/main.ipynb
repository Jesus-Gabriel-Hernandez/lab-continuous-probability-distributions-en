## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python

import scipy.stats as stats

# Function to generate uniformly distributed random numbers using SciPy
def generate_uniform_dist(bottom, ceiling, count):
    return stats.uniform.rvs(loc = bottom, scale = ceiling - bottom, size=count)
```

**Your task:**

        * `count` - how many numbers to generate

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python


import numpy as np
import matplotlib.pyplot as plt

# Function to generate the random numbers for exponentially distributtion
def generate_exponential(mean, size):
    return np.random.exponential(mean, size)

# We generate two sequences
number_seq = generate_exponential(10, 10)
number_seq2 = generate_exponential(10, 100)


plt.hist(number_seq, bins=100, edgecolor='black')
plt.title('Exponential Distribution with size 10')
plt.xlabel('Value')
plt.ylabel('Count')
plt.show()

plt.hist(number_seq2, bins=100, edgecolor='black')
plt.title('Exponential Distribution with size 100')
plt.xlabel('Value')
plt.ylabel('Count')

plt.show()
```

How are the two distributions different?

The mean changes, so the distribution changes as well. 

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e.  = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
# your answer here
# Hint: This is same as saying P(x<15)

# Mean time in the bank
mean_time = 10

calc_time = 15

# We create the rate to use the exponential distribution
rate = 1 / mean_time

# Use the CDF formula to calculate the probability
final_prob = 1 - math.exp(-rate * calc_time)
print(f"The probability for a customer which spends less than {calc_time} minutes would be: {final_prob}")
```

What is the probability that the customer will spend more than 15 minutes

```python


# This is the case for more than 15 minutes
time_above_15m = 1- final_prob
print(f"The probability for a customer which spends more than {calc_time} minutes would be: {time_above_15m}")

```
